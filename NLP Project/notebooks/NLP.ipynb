# Data Acquisition:

This cell loads the dataset, performs sentiment analysis using VADER (from NLTK, good for reviews), and visualizes results. VADER scores polarity from -1 (negative) to +1 (positive).

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from datasets import load_dataset

# Load a sample dataset from HuggingFace
dataset = load_dataset("yelp_review_full", split="train[:1000]")
df = pd.DataFrame(dataset)

# Convert the dataset to match our needs
df = df.rename(columns={'text': 'Text'})
 
# Extract review text (using 'Text' column for full reviews)
reviews = df['Text'].dropna().tolist()[:1000]  

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Analyze sentiment for each review
sentiments = []
for review in reviews:
    score = sia.polarity_scores(review)
    sentiments.append({
        'compound': score['compound'],  
        'pos': score['pos'],  
        'neu': score['neu'],  
        'neg': score['neg']  
    })

# Convert to DataFrame for easier handling
sentiment_df = pd.DataFrame(sentiments)

# Classify overall sentiment
sentiment_df['label'] = sentiment_df['compound'].apply(lambda x: 'positive' if x > 0.05 else ('negative' if x < -0.05 else 'neutral'))

# Summary statistics
print(sentiment_df.describe())
print(f"Sentiment Distribution:\n{sentiment_df['label'].value_counts()}")

# Visualization: Histogram of compound scores
plt.hist(sentiment_df['compound'], bins=20, edgecolor='black')
plt.title('Distribution of Sentiment Scores (Compound)')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()

# Example: Print a few reviews with their sentiments
for i in range(5):
    print(f"Review: {reviews[i][:100]}...")  # Truncate for brevity
    print(f"Sentiment: {sentiment_df.iloc[i]['label']} (Score: {sentiment_df.iloc[i]['compound']:.2f})\n")
```

* Interpretation

Overall Trends: About 51% positive, 31% neutral, 18% negativeindicating a mildly positive skew in the sample. This aligns with Yelp's mix of reviews.

Insights: VADER captures basic sentiment well (e.g., exclamation points boost positivity), but it can miss sarcasm or context. For production use, fine-tune a model on labeled data.
